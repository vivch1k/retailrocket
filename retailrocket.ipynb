```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules, apriori

plt.style.use("ggplot")
```

```python
category_tree = pd.read_csv("data/ecommerce-dataset/category_tree.csv")
events = pd.read_csv("data/ecommerce-dataset/events.csv")
item_properties_part1 = pd.read_csv("data/ecommerce-dataset/item_properties_part1.csv")
item_properties_part2 = pd.read_csv("data/ecommerce-dataset/item_properties_part2.csv")
```

```python
display(category_tree.head(10))
display(events.head(10))
display(item_properties_part1.head(10))
display(item_properties_part2.head(10))
```

```python
df_transaction = events[events["transactionid"].notna()]
df_transaction
```

```python
basket = df_transaction.groupby("transactionid")["itemid"].apply(list)
basket = basket[basket.apply(len) > 1]
basket
```

```python
te = TransactionEncoder()
te_data = te.fit_transform(basket)

df_te = pd.DataFrame(te_data, columns=te.columns_)
```

```python
df_te
```

```python
freq_apriori = apriori(df_te, min_support=0.001, use_colnames=True)
freq_fpgrowth = fpgrowth(df_te, min_support=0.001, use_colnames=True)

rules_apriori = association_rules(freq_apriori, metric="lift", min_threshold=1.2)
rules_fpgrowth = association_rules(freq_fpgrowth, metric="lift", min_threshold=1.2)
```
